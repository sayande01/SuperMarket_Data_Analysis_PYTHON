```python
df['City'].value_counts()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df.head()
```

```python
df['Month'].value_counts()
```

```python
# Define a dictionary mapping numeric values to month names
month_mapping = {1: 'Jan', 2: 'Feb', 3: 'March'}

# Replace numeric values with month names
df['Month_Name'] = df['Month'].replace(month_mapping)
```

```python
df.head()
```

```python
df3  =  df.groupby('Month_Name')['gross income'].sum()
df3
```

```python
# Sample DataFrame
data = {'Month': df3.index,
        'Value': df3.values}
df4 = pd.DataFrame(data)
```

```python
df4
```

## For sns.lineplot we need to create a dataframe from the pandas series

```python
sns.lineplot(data = df4, x = 'Month', y = 'Value', marker='o', color='b' )
```

```python
# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(df3.index, df3.values, marker='o', color='b', linestyle='-')

# Adding title and labels
plt.title('Month-wise Gross Income')
plt.xlabel('Month')
plt.ylabel('Total Sales')

# Adding grid lines
plt.grid(True)

# Display the line chart
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data = df, x='Rating', bins=20, kde=True, hue = 'Gender')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Histogram of Rating')
plt.show()
```

```python
# Heatmap of passenger counts
pivot_table = df.pivot_table(index='Product line', columns='Payment', values='Quantity', aggfunc=np.mean)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='Greens', annot=True, linewidths=0.5)
plt.xlabel('Payment type')
plt.ylabel('Category')
plt.title('Heatmap of Avg Quantities Sold by Different Payment types')
plt.show()
```

```python
df4 = df.groupby('City')['gross income'].sum()
df4
```

```python
plt.figure(figsize=(10, 6))
colors = ['skyblue', 'salmon', 'lightgreen']
plt.barh(df4.index, df4.values, color=colors)

# Adding title and labels
plt.title('City-wise Total Sales')
plt.ylabel('City')
plt.xlabel('Total Sales')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the bar chart
plt.show()
```

```python
# Calculate values for the donut chart
city_counts = df['City'].value_counts()
labels = city_counts.index
sizes = city_counts.values

# Create a donut chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'], wedgeprops=dict(width=0.4))
plt.title('City wise Customer distribution')
plt.axis('equal')
plt.show()
```
